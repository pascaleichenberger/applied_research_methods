# Combining & organizing data

## Libraries and settings

```python
# Libraries
import os
import pandas as pd
import numpy as np
import fnmatch

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Show current working directory
print(os.getcwd())
```

## Importing the prepared rental apartments data

```python
# Read data to pandas data frame
df_orig = pd.read_csv('./Data/apartments_data_prepared.csv', 
                      sep=',', 
                      encoding='utf-8')

# Copy of data with selected colums
columns = [ 'web-scraper-order', 
            'address_raw',
            'rooms', 
            'area', 
            'luxurious',
            'price',
            'price_per_m2']
df = df_orig[columns]

# Get number of rows and columns
print(df.shape)

# Show first records
df.head(5)
```

## Combining data from different sources

### Reading rental apartment data with geocoded addresses

```python
# Meaning of variables
# lat: geographical latitude
# lon: geographical longitude
# bfs_number: official municipality id
# bfs_name: official municipality name

# Geocoded data (i.e. data with latitude and longitude)
df_geo = pd.read_csv('./Data/apartments_data_geocoded.csv', 
                     sep=',', 
                     encoding='utf-8')

# Show data
df_geo[['web-scraper-order',
        'address_raw',
        'lat', 
        'lon', 
        'bfs_number', 
        'bfs_name']].head()
```

### Join geo-information to rental apartment data using .merge()

```python
df2 = df.merge(df_geo[['web-scraper-order', 
                       'lat', 
                       'lon', 
                       'bfs_number', 
                       'bfs_name']], 
               on="web-scraper-order")
df2.head()
```

### Reading municipality-level data

```python
# Meaning of variables:
# bfs_number: official municipality id
# bfs_name: official municipality name
# pop: number of residents (=population)
# pop_dens: population density (pop per km2)
# frg_pct: percentage foreigners
# emp: numer of employees
# mean_taxable_income: mean taxable income

df_municip = pd.read_excel('./Data/municipality_data_with_taxable_income.xlsx', 
                           sheet_name='data_for_import')
df_municip.head(5)
```

### Join municipality data to rental apartment data using .merge()

```python
# Merge needs a key which must be identical in both data sets (here the key is 'bfs_number')
df3 = df2.merge(df_municip[['bfs_number', 
                            'pop', 
                            'pop_dens', 
                            'frg_pct', 
                            'emp',
                            'mean_taxable_income']], 
                on="bfs_number")
df3.head(5)
```

### Export data to file

```python
# Remove missing values which may have ocured from the merging process
df3 = df3.dropna()

# Check for missing values per column
print(df3.isna().sum().sum())

# Remove duplicates
df3 = df3.drop_duplicates()

# Check for duplicates
print(df3.duplicated().sum().sum())

# Count number of rows and columns
print(df3.shape)

# Export apartment data to file
df3.to_csv('./Data/apartments_data_enriched.csv',
            sep=',',
            encoding='utf-8')
```

#To determine if there is a relationship between the variable area and price_per_m2, we can perform a correlation analysis. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'area': [50, 75, 100, 125, 150],
    'price_per_m2': [2000, 1800, 1600, 1400, 1200]
}

df = pd.DataFrame(data)

# Calculate correlation
correlation = df['area'].corr(df['price_per_m2'])
print(f"Correlation between area and price_per_m2: {correlation}")

# Plot the relationship
sns.scatterplot(x='area', y='price_per_m2', data=df)
plt.title('Relationship between Area and Price per m2')
plt.xlabel('Area (m2)')
plt.ylabel('Price per m2')
plt.show()
```

#Explanation

The corr method calculates the Pearson correlation coefficient between area and price_per_m2. A negative correlation indicates that as the area increases, the price per square meter tends to decrease.
The scatter plot visualizes the relationship between area and price_per_m2. If the points form a downward trend, it suggests a negative relationship.

A negative correlation between area and price_per_m2 suggests that larger apartments tend to have a lower price per square meter. This could be due to economies of scale, where larger apartments are relatively cheaper per unit area compared to smaller ones.
If the correlation were positive, it would suggest that larger apartments are more expensive per square meter, possibly due to higher demand for larger living spaces.
